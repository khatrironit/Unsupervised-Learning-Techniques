```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
```

```python
X,Y = make_moons(n_samples = 250,noise = 0.05)
```

```python
plt.scatter(X[:,0],X[:,1],c=Y)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 2)
kmeans.fit(X)
```

```python
centers = kmeans.cluster_centers_
label = kmeans.labels_

plt.scatter(X[:,0],X[:,1],c=label)
plt.scatter(centers[:,0],centers[:1],marker = "*",color = "orange")
plt.show()
```

# DBSCAN
        - Density Based Spatial Clustering of Application with Noise
        - Based on the idea that a cluster is a high density area surrounded by low density region.
        - Starts by exploring a small area, if density is good enough it is considered as a part of the cluster and explores  neighbours and so on, so it groups regions baesd on density.

```python
from sklearn.cluster import DBSCAN
```

```python
dbs = DBSCAN(eps = 0.3,min_samples = 5)
```

```python
dbs.fit(X)
```

```python
label1 = dbs.fit_predict(X)
```

```python
plt.scatter(X[:,0],X[:,1],c = label1)
plt.show()
```
